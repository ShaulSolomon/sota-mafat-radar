create a credentials.colab.ini file with these parameters: NGROK_TOKEN, NGROK_PASS

```python
import configparser
import os.path
from os import path

creds_path_ar = ["../credentials.ini","credentials.colab.ini"]
ngrokToken = ""
password = ""

for creds_path in creds_path_ar:
  if path.exists(creds_path):
      config_parser = configparser.ConfigParser()
      config_parser.read(creds_path)
      ngrokToken = config_parser['MAIN']["NGROK_TOKEN"]
      password = config_parser['MAIN']["NGROK_PASS"]
      break
```
